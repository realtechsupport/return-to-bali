```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imgL = cv2.imread("../collection/category/0000.jpg", 0)
imgR = cv2.imread("../collection/category/0039.jpg", 0)
plt.figure(figsize=(20,20), dpi = 90)
plt.subplot(121)
plt.imshow(imgL,'gray')
plt.subplot(122)
plt.imshow(imgR,'gray')
plt.show()
```

```python
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=11)
disparity = stereo.compute(imgL,imgR)
plt.figure(figsize=(12,12), dpi = 70)
plt.imshow(disparity,'gray')
plt.show()
```
